```python
#create function to do this cleaning for all dataframes 2012-2018
def clean_income(dataframe, median_USnatl, year):
    '''takes raw median income data, cleans it and shifts to match the national median US income '''
    header_row = 0
    dataframe.columns = dataframe.iloc[0]
    dataframe.drop(0, inplace=True)
    dataframe['Median income (dollars)!!Estimate!!Households'].replace('-', np.nan, inplace=True)
    dataframe['Median income (dollars)!!Estimate!!Households'].replace('250,000+', '250000', inplace=True)
    dataframe['Median income (dollars)!!Estimate!!Households'].replace("2,500-", "2500", inplace=True)
    dataframe['Median income (dollars)!!Estimate!!Households'] = dataframe['Median income (dollars)!!Estimate!!Households'].astype('float')
    dataframe['Median income (dollars)!!Estimate!!Households'].fillna(0, inplace=True)
    median_shift = median_USnatl - np.quantile(dataframe['Median income (dollars)!!Estimate!!Households'], 0.5)
    dataframe['Median income (dollars)!!Estimate!!Households'] = dataframe['Median income (dollars)!!Estimate!!Households'] + median_shift
    dataframe['Year'] = year
    dataframe2 = dataframe[['Year', 'Median income (dollars)!!Estimate!!Households']]
    dataframe2.columns = ['Year', 'Median_Income$']
    return dataframe2;
```
