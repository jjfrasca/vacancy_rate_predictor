### clean zipcode level data

```python
#convert dtype object to float
df_2011[['Estimate!!Total', 'Margin of Error!!Total', 'Estimate!!Total!!Occupied', 'Margin of Error!!Total!!Occupied', 'Estimate!!Total!!Vacant', 'Margin of Error!!Total!!Vacant']] = df_2011[['Estimate!!Total', 'Margin of Error!!Total', 'Estimate!!Total!!Occupied', 'Margin of Error!!Total!!Occupied', 'Estimate!!Total!!Vacant', 'Margin of Error!!Total!!Vacant']].astype('float')
df_2011.dtypes
```

```python
#calculate vacany rate, adding new column
df_2011['Vacancy_Rate%'] = df_2011['Estimate!!Total!!Vacant']/df_2011['Estimate!!Total']*100
df_2011.head()
```

```python
#calculate margin of error of vacancy rate (using error propogation formula <https://www.statisticshowto.com/error-propogation/>)
df_2011['MOE-VacancyRate%'] = np.sqrt((df_2011['Margin of Error!!Total!!Vacant']/df_2011['Estimate!!Total!!Vacant'])**2 + (df_2011['Margin of Error!!Total']/df_2011['Estimate!!Total'])**2) * (df_2011['Estimate!!Total!!Vacant']/df_2011['Estimate!!Total'])*100
df_2011.head()
```

```python
#do cleaning on all the remaining data frames
DFs = [df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018]

for DF in DFs:  
    DF.columns = DF.iloc[0]
    DF.drop(0, inplace=True)
    DF[['Estimate!!Total', 'Margin of Error!!Total', 'Estimate!!Total!!Occupied', 'Margin of Error!!Total!!Occupied', 'Estimate!!Total!!Vacant', 'Margin of Error!!Total!!Vacant']] = DF[['Estimate!!Total', 'Margin of Error!!Total', 'Estimate!!Total!!Occupied', 'Margin of Error!!Total!!Occupied', 'Estimate!!Total!!Vacant', 'Margin of Error!!Total!!Vacant']].astype('float')
```

```python
for DF in DFs:
    DF['Vacancy_Rate%'] = DF['Estimate!!Total!!Vacant']/DF['Estimate!!Total']*100
    DF['MOE-VacancyRate%'] = np.sqrt((DF['Margin of Error!!Total!!Vacant']/DF['Estimate!!Total!!Vacant'])**2 + (DF['Margin of Error!!Total']/DF['Estimate!!Total'])**2) * (DF['Estimate!!Total!!Vacant']/DF['Estimate!!Total'])*100
    DF.loc[DF['MOE-VacancyRate%'] > DF['Vacancy_Rate%'], 'MOE-VacancyRate%'] = DF['Vacancy_Rate%']
    DF.drop(['Estimate!!Total', 'Margin of Error!!Total', 'Estimate!!Total!!Occupied', 'Margin of Error!!Total!!Occupied', 'Estimate!!Total!!Vacant', 'Margin of Error!!Total!!Vacant'], axis=1, inplace=True)
```

```python
for DF in DFs:
    DF.fillna(0, inplace=True)
```
